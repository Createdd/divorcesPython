```python
data = pd.read_csv('divorce.csv', delimiter=";")
data
```

```python
year = data.REF_YEAR
total = data.TOTAL
husband = data.HUSBAND_RESPONSIBLE
wife = data.WIFE_RESPONSIBLE
both = data.BOTH_RESPONSIBLE
neither = data.NEITHER_RESPONSIBLE
plt.title("Responsibilities of divorces over the years")
plt.plot(year, total)
plt.plot(year, husband)
plt.plot(year, wife)
plt.plot(year, both)
plt.plot(year, neither)
plt.legend(['total', 'only husband', 'only wife', 'both', 'neither'])
plt.show()
```
